```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
sns.set(style="ticks")
```

```python
from sklearn.datasets import load_wine
```

```python
data, tar= load_wine(return_X_y=1, as_frame=1)
```

```python
fig, ax = plt.subplots(figsize=(18, 18))
sns.heatmap(data.corr(), annot = True)
```

У целевого признака наблюдается достаточно сильная корелляция с признаками flavanoids (-0,85), и od280/od315_of_diluted_wines (-0,79). Но стоит заметить, что это два признака также достаточно сильно кореллируют между собой (0,79), исходя из чего один из них можно удалить (хотя решение спорное, всё же корелляция не слишком высокая)
Также мы можем отметить, что присутствует сильная корелляция у признаков flavanoids и total_phenols (0,86), на основании чего можно задуматься о удалении одного из признаков из выборки.
